```python
import pandas as pd
data = pd.read_csv('../data/monthly_county_losses_sub.csv')
df = pd.DataFrame(data)
```

```python
#number one
df = df.set_index(df.county_name) #setting index = county name
df[:5]
```

```python
#number two
tornados = df.hazard_code_desc == 'Tornado'
tornado_kills = df[tornados].fatalities.sum()               #how many people killed by tornado
print("Tornado:",tornado_kills)
hurrtrop = df.hazard_code_desc == 'Hurricane/Tropical Storm'
hurrtrop_kills = df[hurrtrop].fatalities.sum()              #how many people killed by hurricane/tropical storm
print("Hurricane/Tropical storm:",hurrtrop_kills)
```

```python
#number three
import numpy as np
hazard_codes = ['Avalanche', 'Coastal', 'Drought', 'Earthquake', 'Flooding', 'Fog', 'Hail', 'Heat', 'Hurricane/Tropical Storm', 'Landslide',
 'Lightning', 'Severe Storm/Thunder Storm', 'Tornado', 'Tsunami/Seiche', 'Volcano', 'Wildfire', 'Wind', 'Winter Weather']

sums = []
big = 0
for kind in hazard_codes:
    if df[df.hazard_code_desc==kind].fatalities.sum() > big:
        big = df[df.hazard_code_desc==kind].fatalities.sum()   #choosing the largest number of fatalities
    sums.append(df[df.hazard_code_desc==kind].fatalities.sum())
print(hazard_codes[sums.index(big)]) #printing the deadliest type
```

```python
#number four
bvtorns = (df.hazard_code_desc == 'Tornado')&(df.county_name == 'Buena Vista') #choosing tornados in BV county
median = df[bvtorns].property_damage.median()
print("Median:",median)
mean  = df[bvtorns].property_damage.mean()
print("Mean:",mean)
```

```python
#number five
df['property_damage_2017'] = df.property_damage * 1.03  #1.03 is the conversion rate
df['crop_damage_2017'] = df.crop_damage * 1.03

df[:5]
```

```python
#number six
df['fatalities_per_capita'] = df.fatalities/df.population
df['injuries_per_capita'] = df.injuries/df.population     
df[:5]
```

```python
#number seven
df = df.sort_values(by = 'fatalities_per_capita',ascending = False)   
df[:5]
```

```python
#number eight
fatinj = df['fatalities'].corr(df['injuries'])
fatprop = df['fatalities'].corr(df['property_damage'])  #correlation coefficient between variables using pandas
print("Fatalities vs injuries:",fatinj)
print("Fatalities vs propery damage:",fatprop)
```

```python
#number nine
tornados = df.hazard_code_desc == 'Tornado'
tornfatinj = df[tornados].fatalities.corr(df[tornados].injuries)            #correlation except only for tornado incidents
print("Tornados only for fatalities vs injuries:",tornfatinj)
tornfatprop = df[tornados].fatalities.corr(df[tornados].property_damage)
print("Tornados only for fatalities vs property damage:",tornfatprop)
```
